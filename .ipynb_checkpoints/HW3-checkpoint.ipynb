```python

import math as ma
import numpy as np
import matplotlib.pyplot as plt
## графики
x = np.linspace(-5, 5, 25)
y = 1/1+(ma.e)**(-x)
u = x**4
fig, ax = plt.subplots()
##линии
ax.plot(x, y + 100,
        linestyle = '-',
        linewidth = 2,
        color = 'darkmagenta')
ax.plot(x, y +200,
        linestyle = '-',
        linewidth = 2,
        color = 'r')
ax.plot(x, u,
        linestyle = '-',
        linewidth = 2,
        color = 'y')
##маркеры
ax.plot(u,
        marker = 'o',
        color = 'seagreen')
ax.plot(y,
        marker = 'o',
        markersize = 2,
        color = 'lime',
        linewidth = 1,
        markerfacecolor = 'lawngreen',
        markeredgecolor = 'limegreen',
        markeredgewidth = 10)
ax.plot(x, y,)
ax.grid()

##вывод
plt.show()
```
